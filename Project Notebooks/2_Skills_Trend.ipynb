```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_PH = df[(df['job_location'] == "Philippines") & (df['job_title_short'] == "Data Analyst")]

df_DA_PH
```

```python
df_DA_PH['job_posted_month_no'] = df_DA_PH['job_posted_date'].dt.month

df_DA_PH
```

```python
df_DA_PH_exploded = df_DA_PH.explode("job_skills")
```

```python
df_DA_PH_exploded2 = df_DA_PH_exploded.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)

df_DA_PH_exploded2
```

```python
df_DA_PH_exploded2.loc['Total'] = df_DA_PH_exploded2.sum()
```

```python
df_DA_PH_exploded2
```

```python
df_DA_PH_exploded2 = df_DA_PH_exploded2[df_DA_PH_exploded2.loc['Total'].sort_values(ascending=False).index]

df_DA_PH_exploded2
```

```python
df_DA_PH_exploded2
```

```python
df_DA_PH_exploded2.index
```

```python
df_DA_PH_exploded2.drop("Total", axis=0, inplace=True)

df_DA_PH_exploded2
```

```python
df_totals = df_DA_PH.groupby("job_posted_month_no").size()
```

```python
df_percentages = df_DA_PH_exploded2.div(df_totals/100, axis="index")

df_percentages
```

```python
df_percentages.reset_index(inplace=True)

df_percentages

#df_percentages.set_index("job_posted_month")
```

```python
df_percentages['job_posted_month_no'] = pd.to_datetime(df_percentages['job_posted_month_no'], format='%m').dt.strftime('%b')

df_percentages.rename(columns={'job_posted_month_no':'job_posted_month'}, inplace=True)

df_percentages
```

```python
df_percentages.set_index("job_posted_month", inplace=True)
```

```python
df_percentages
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
top_no = 5

df_plot = df_percentages.iloc[: , :top_no]


ax = sns.lineplot(df_plot, dashes=False, palette= "tab10")
sns.despine()
plt.xlabel('Month')
plt.ylabel('Likelihood')
plt.legend().remove()
formatter = FuncFormatter(lambda x, _: f'{int(x)}%')
ax.yaxis.set_major_formatter(formatter)
plt.suptitle(f'Likelihood of the Top {top_no} Skills to be Demanded for Each Month in the Philippines', size=15)


plt.text(11.2, 50.1, df_percentages.columns[0])
plt.text(11.2, 33.2, df_percentages.columns[1])
plt.text(11.2, 23, df_percentages.columns[2])
plt.text(11.2, 25, df_percentages.columns[3])
plt.text(11.2, 26.9, df_percentages.columns[4])

plt.tight_layout()
plt.show()
```
